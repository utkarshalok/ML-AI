```python
import numpy as np
import pylab
import scipy.stats as stats
```

```python
std_normal=np.random.normal(loc=0,scale=1,size=1000)
for i in range(0,101):
    print(i,np.percentile(std_normal,i))
```

Plotting of **QQ-Plot** <br> 
**Uses** : <br>
To check whether two distributions are same <br>
To check whether a function follows Normal distribution <br>
Normal/Gaussian Distribution with **Size**=1000 **Mean**=20 and **Variance**=5

```python
# N(20,5)
measurements=np.random.normal(loc=20,scale=5,size=1000)
stats.probplot(measurements,dist='norm',plot=pylab)
pylab.show()
```

Now we will check with **Uniform Distribution** with respect to **Normal Distribution**

```python
m=np.random.uniform(low=-1,high=1,size=10000)
stats.probplot(m,dist="norm",plot=pylab)
pylab.show()
```

**Log-Normal Distribution**

```python
import matplotlib.pyplot as plt
mu,sigma=3,1
s=np.random.lognormal(mu,sigma,1000)
count, bins, ignored = plt.hist(s, 100, normed=True, align='mid')
x = np.linspace(min(bins), max(bins), 1000)
pdf = (np.exp(-(np.log(x) - mu)**2 / (2 * sigma**2)) / (x * sigma * np.sqrt(2 * np.pi)))
plt.plot(s,pdf)
plt.show()
```

<h1>How to Randomly Sample Data Points</h>

```python
from sklearn import datasets
iris=datasets.load_iris()
ir=iris.data
ir.shape
```

Sample of 30 points from 150 points randomly

```python
sampled_data=[]
for i in range(0,150):
    if random.random() <= 0.2:
        sampled_data.append(ir[i:])
    if len(sampled_data)==30:
        break
len(sampled_data)
```
