```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Number of points
n=1000
#Distance or Time Period
Lx=100
#Angular frequency :  ω = 2πf 2π/T
omega= 2*np.pi/Lx

x=np.linspace(0,Lx,n)

#Sine wave
wave1= 10.0* np.sin(5.0* omega*x)

plt.figure(1)
plt.plot(x,wave1,color='blue')
plt.show()
```

```python
wave2= 5.0* np.sin(20.0* omega*x)
wave3= 0.5* np.sin(1.0* omega*x)
#Final Wave
wave=wave1+wave2+wave3
plt.figure(1)
plt.title("Composite wave signal")
plt.plot(x,wave,color='blue')
plt.show()
```

**Numpy fft value calculates it multiplied by number of point(n) and it is seperated between negative and positive values
So, Actual fft values is multiplied by 2 . **

```python
#Creating all necessary frequencies
freqs=np.fft.fftfreq(n)
#print(freqs)
fft_value=np.fft.fft(wave)
#print(fft_value)

fft_actual=2.0*np.abs(fft_value/n)
#print(fft_actual)
```

```python
mask=freqs>0

plt.plot(freqs[mask],fft_actual[mask],color='red')
plt.show()
```
