```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
haberman=pd.read_csv("haberman.csv")
haberman.columns
```

```python
haberman.isnull().values.any()
```

So we saw we do not have any **null** values in the dataframe

```python
haberman.describe()
```

```python
print(np.mean(haberman["age"]))
print(np.median(haberman["age"]))
print(np.percentile(haberman["age"],np.arange(0,100,25)))
print(np.percentile(haberman["age"],[25,75,100]))
```

As **mean** and **median** is same we can say that there is no outliers in age column . 

```python
haberman["status"].value_counts()
```

```python
sns.pairplot(haberman,hue="status",size=5)
plt.show()
```
